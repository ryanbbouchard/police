# DataCamp: Analyzing Police Activity with Pandas

```python
import pandas as pd
```

```python
ri = pd.read_csv('data/police_ri.csv')
ri.head(3)
```

```python
# look at the shape of the data
ri.shape
```

We see that there are 31 columns with 509,681 rows. 

Next, let's check for the presence of null values and clean up the data

```python
# locate the missing values (1) - visual inspection
ri.isnull()
```

```python
# locate missing values (2) - sum up the number of missing values per column. 
ri.isnull().sum().sort_values()
```

We see that there are columns were nearly all values are null.

Let's drop 'raw_SearchResultThree' completely and then check the new shape of the data.

```python
ri.drop('raw_SearchResultThree', axis='columns',inplace=True)

ri.shape
```

Next, let's drop rows where some critical values, such as search time, are missing.

```python
ri.dropna(subset=['date', 'time'], inplace=True)

ri.shape
```
